**Step -1 : Reading Image**

```python
img = cv2.imread('image_2.jpg') # reading image
```

```python
img.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img)
plt.show()
```

```python
r, g, b  = cv2.split(img)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(131)
plt.imshow(r, cmap ='gray')
plt.title('Red Region')

plt.subplot(132)
plt.imshow(g, cmap ='gray')
plt.title('Green Region')

plt.subplot(133)
plt.imshow(b, cmap ='gray')
plt.title('Blue Region')
plt.show()

plt.imshow(img)
plt.title('Blue Green Red Region')
plt.show()
```

```python
img_new = cv2.merge((b,g,r)) # merging
```

```python
plt.figure(figsize=(15,8))
plt.subplot(131)
plt.imshow(r, cmap ='gray')
plt.title('Blue Region')

plt.subplot(132)
plt.imshow(g, cmap ='gray')
plt.title('Gree Region')

plt.subplot(133)
plt.imshow(b, cmap ='gray')
plt.title('Red Region')
plt.show()

plt.imshow(img_new)
plt.title('Blue Green Red Region')
plt.show()
```

**Step - 2: Converting into Grayscale**

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray,cmap = 'gray')
plt.show()
```

```python
cv2.imwrite('grayimag.png',gray)
```

```python
gray.min(), gray.max()
```

**Step - 3 : Thresholding**

```python
ret, thresh1 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(thresh1,cmap='gray')
plt.title('BINARY')
plt.subplot(122)
plt.imshow(thresh2,cmap='gray')
plt.title('BINARY INV')
plt.show()
```
