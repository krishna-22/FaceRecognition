### Beyond Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Number2.jpg')
```

**Visualize picture**

```python
b, g ,r = cv2.split(img)
img = cv2.merge((r,g,b))
plt.imshow(img)
plt.show()
```

## Coverting into gray scale

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

###  Resize the image

```python
img_re = cv2.resize(gray,(12,12))
```

```python
img_re
```

### Thresholding

- Binary image

Lets convert values more than 127 with 255
and less than 127 with 0

```python
img_re[img_re > 127] = 255
img_re[img_re <= 127] = 0
```

```python
img_re
```

```python
thresh = img_re/255.0
thresh
```

```python
plt.imshow(thresh,cmap='gray')
```

### Reshape into row matrix or column 

```python
the_re = thresh.reshape(1,-1)
the_re
```

```python
fig = plt.figure(figsize=(10,15))
plt.imshow(the_re,cmap ='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
fig.savefig('barcode.png')
```
